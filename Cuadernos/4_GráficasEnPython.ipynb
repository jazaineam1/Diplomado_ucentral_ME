# Introducción a Seaborn

Seaborn es una libreria de python para la visualiazación de datos basada en matplotlib. Esta librería esta hecha para construir gráficos estadísticos y tiene la ventaja de estar estrechamente integrada a la librería de pandas. 

Como siempre, antes de comenzar a producir cualquier gráfico, es necesario tener un amplio entendimiento de la estructura de los datos, así como también, tener el conjunto de datos preprocesado y listo para trabajar. Dentro de las funciones que ofrece Seaborn se destacan las siguientes:

- Esta orientada al trabajo con dataframes de pandas, lo que permite examminar relaciones entre múltiples variables
- Tiene un gran soporte para el manejo de variables categóricas y para pintar observaciones y estadísticas agregadas
- Opciones para visualizar distribuciones univaridas y bivariadas
- Estimación automática como por ejemplo el trazado de modelos de regresión lineal
- Vistas sobre la estructura general de conjuntos de datos
- Abstracciones de alto nivel para estructurar cuadrículas de múltiples gráficos
- Control del estilo con varios temas incorporados
- Herramientas para elegir paletas de colores 

La pregunta fundamental es: __¿Cómo se pueden visualizar datos de alta dimensionalidad de manera apropiada?__

##Tips para creación de gráficos efectivos

- Al momento de crear un conjunto de gráficos que presentan algún fenómeno o situación particular, siempre se debe pensar en contar una __historia__. 
- Tenga en cuenta el contexto y la audiencia que va a recibir esta información
- Seleccionar el gráfico ideal, el que sea más efectivo en cada ocasión
- El desorden es el enemigo, muchas veces menos es más
- Enfocar la atención de la audiencia
- A la hora de cosntruir el gráfico piense como un diseñador para comunicar con datos

## Gráficas de dispersión

Las gráficas de dispersión se recomiendan para evidenciar relaciones entre variables y para mostrar la distribución de variables que generalmente son numéricas. En general, este es un buen primer paso para observar relaciones y distribución de los datos. 

Seaborn dispone de la función __relplot__, la cual presenta variadas e interesantes parametrizaciones.

### Ejemplo 1
> Iniciamos cruzando dos variables numéricas usando $x, y$, como ejemplo tomamos el dataframe Iris:


```python
# importar la libreria de Seaborn
sns.relplot(x="petal.length", y="sepal.length", data=dfi)
```




````python
# Celda de código para probar
sns.relplot(x="petal_length", y="sepal_length", data=dfi)
````

### Otras opciones de visualización de gráficas dispersas

Varias dimensiones se pueden visualizar en una gráfica de dispersión
- $x$ variable numérica (puede tener múltiples valores)
- $y$ variable numérica
- $hue$ variable categórica (agrupa)
- $size$ diferentes tamaños de la burbuja, variable numérica 
- $style$ pone marcas diferentes, variable categórica (agrupa)
- $col$ organiza filtros por columna, variable categórica (filtra)
- $row$ organiza filtros por fila, variable categórica (filtra)

## Gráficos de líneas

Generalmente cuando se tienen una o muchas variable numéricas que dependen de una variable independiente y ordenada (como por ejemplo el tiempo), lo ideal es realizar gráficos de líneas. En estos gráficos se pueden representar y comparar fácilmente varias series de datos, así como también mostrar comportamientos asociados a la variable independiente.

Cuando se tienen múltiples valores de la serie por cada valor de la varaiable independiente, se  gráfica de manera automática el intervalo de confianza del 95% calculado usando el método de bootstrapping. 



```python
# intervalo de confianza desabilidtado
sns.relplot(x="nota", y="prom", hue="jornada", ci=False, kind='line',data=dfx)
# intervalo de confianza habilitado
sns.relplot(x="nota", y="prom", hue="jornada", kind='line',data=dfx)
```



````python
sns.relplot(x="nota", y="prom", hue="jornada", ci=False, kind='line',data=dfx)
sns.relplot(x="nota", y="prom", hue="jornada", kind='line',data=dfx)
````
