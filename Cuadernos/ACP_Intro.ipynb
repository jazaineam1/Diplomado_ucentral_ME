# Análisis de componentes principales

## Reducción de Dimensionalidad

La reducción de dimensionalidad es un proceso fundamental en el análisis de datos que nos permite simplificar y representar datos complejos en un espacio de menor dimensión. En este capítulo, exploraremos en detalle las técnicas y conceptos clave relacionados con la reducción de dimensionalidad.

## ¿Por qué es necesaria la reducción de dimensionalidad?

En muchos conjuntos de datos, especialmente en aquellos con un gran número de variables, puede haber redundancia, características irrelevantes o variables altamente correlacionadas. Estas situaciones pueden dificultar el análisis y la interpretación de los datos, así como aumentar la complejidad computacional. La reducción de dimensionalidad nos ayuda a abordar estos desafíos al seleccionar un subconjunto de variables más representativo y significativo.

## Técnicas de reducción de dimensionalidad: selección y extracción de características

Existen dos enfoques principales para la reducción de dimensionalidad: selección de características y extracción de características.

- Selección de características: En este enfoque, seleccionamos un subconjunto de características o variables del conjunto de datos original. Esto implica elegir las variables más relevantes o informativas y descartar las irrelevantes. Algunos métodos populares de selección de características son el filtro estadístico, la selección basada en modelos y la eliminación recursiva de características.

- Extracción de características: En contraste con la selección de características, la extracción de características crea nuevas variables que son combinaciones lineales de las variables originales. Este enfoque busca proyectar los datos en un nuevo espacio de menor dimensión que conserve la mayor cantidad posible de información relevante. Un ejemplo común de extracción de características es el Análisis de Componentes Principales (PCA), que transforma los datos en un conjunto de componentes ortogonales ordenados por su varianza.

## Análisis de Componentes Principales (PCA): teoría y aplicación

El Análisis de Componentes Principales (PCA) es una de las técnicas más utilizadas para la reducción de dimensionalidad. PCA busca encontrar una transformación lineal de los datos originales para proyectarlos en un espacio de menor dimensión mientras se mantiene la mayor varianza posible.

En la aplicación de PCA, se siguen los siguientes pasos:

1. Estandarización de los datos: Es importante asegurarse de que todas las variables tengan la misma escala antes de aplicar PCA.

2. Cálculo de la matriz de covarianza o de correlación: Se calcula la matriz de covarianza si las variables están en diferentes escalas, o la matriz de correlación si las variables están en la misma escala.

3. Cálculo de los autovectores y autovalores: Se calculan los autovectores y autovalores de la matriz de covarianza o de correlación. Los autovectores representan las direcciones principales de variabilidad en los datos, y los autovalores indican la cantidad de varianza explicada por cada autovector.

4. Selección de componentes principales: Se seleccionan las componentes principales con los autovalores más altos. Estas componentes representan las nuevas variables que capturan la mayor cantidad de información relevante del conjunto de datos original.

## Otros métodos de reducción de dimensionalidad

Además del PCA, existen otros métodos populares de reducción de dimensionalidad que pueden ser útiles en diferentes situaciones:

- Análisis de Discriminante Lineal (LDA): Se utiliza en problemas de clasificación para encontrar una combinación lineal de características que maximice la separación entre clases.

- t-SNE (t-Distributed Stochastic Neighbor Embedding): Es una técnica no lineal de reducción de dimensionalidad que se utiliza comúnmente para visualización. Preserva las relaciones de proximidad entre los puntos en el espacio original al proyectarlos en un espacio de menor dimensión.

- Factorización de matrices: Es un enfoque que descompone la matriz de datos original en dos o más matrices de menor rango. Algunos métodos populares son la Factorización de Matrices no Negativas (NMF) y la Descomposición en Valores Singulares (SVD).

## Evaluación de la reducción de dimensionalidad

Es esencial evaluar la calidad de la reducción de dimensionalidad realizada. Algunas métricas comunes utilizadas para evaluar los resultados son:

- Varianza explicada: Indica cuánta varianza del conjunto de datos original es explicada por las componentes seleccionadas.

- Preservación de la estructura: Evalúa si las relaciones y la estructura de los datos originales se conservan en el espacio de menor dimensión.

- Evaluación del rendimiento del modelo: Se evalúa cómo la reducción de dimensionalidad afecta el rendimiento de un modelo predictivo o clasificador aplicado a los datos reducidos.

En este capítulo, hemos explorado los fundamentos de la reducción de dimensionalidad, incluidos los conceptos clave, las técnicas de selección y extracción de características, así como el Análisis de Componentes Principales (PCA) y otros métodos populares. Estas técnicas nos permiten simplificar y representar los datos de manera más efectiva, facilitando el análisis y la interpretación en diversos campos del análisis de datos.


## Introducción




> * El análisis en componentes principales es una técnica de análisis estadístico multivariante que se clasifica entre los métodos de simplificación o reducción de la dimensión. Se aplica cuando se dispone de un conjunto elevado de variables y busca reescribir los valores a partir de combinaciones lineales que se denominan componentes principales o factores.
>
> *  Es un método que permite describir, de un modo
sintético, la estructura y las interrelaciones de las variables originales en el fenómeno que se estudia a partir de las componentes obtenidas que, naturalmente, habrá que interpretar y «nombrar».
>
>* La reducción de variables a pocas componentes puede simplificar la aplicación sobre otras técnicas multivariantes (regresión, clusters, etc.)

Obtenemos las componentes principales a partir de la matriz de covarianzas o correlaciones y sus valores y vectores propios. Como dicha matriz es semidefinida positiva grantizamos la nonegatividad de los valores propios. Los organizamos de mayor a menor para definir el orden de los componentes.



## Objetivo operativo del ACP

* Describir la variación de un conjunto de variables correlacionadas, $x^T=(x_1,x_2,\cdots,x_q)$ en  términos de unas nuevas variables NO correlacionadas, $y^T=(y_1,y_2,\cdots,y_q)$, las cuales sean combinaciones lineales de las variables $x$.



Para lograr este objetivo iniciamos tipificando las variables, es decir, hacemos una transformación simple que produzca unas nuevas variables $Z_i$ con media $0$ y desviación estándar $1$.

Lo hacemos porque, buscamos una matriz $U$ de $q\times q$ tal que:

$Y=Z\cdot U$

con $Z$ la matriz de todas las observaciones tipificadas.



**Pregunta**

Si $Z$ es la matriz $n\times q$ que contiene todas las observaciones tipificadas ¿Qué puede decir de la matriz $Z^T Z$ de tamaño $q\times q$?

**Tipificación o estandarización**
Para usar la matriz de covarianzas:

$$x^*=x-\bar{X}$$

Para usar la matriz de correlaciones:

$$x^*=\cfrac{x-\bar{X}}{s}$$

**Observación importante**

Si $X$  es una matriz de tamaño $n\times k$ que simboliza una base de datos de $n$ datos y $k$ variables, entonces su matriz de covarianza es:
$$\cfrac{1}{n-1}(X^*)^T\cdot X^*$$
donde $X^*$ es ta tipificación con el promedio ($x_i^*=x_i-\bar{X}$), y la matriz de correlaciones es:
$$\cfrac{1}{n-1}(X^*)^T\cdot X^*$$
donde $X^*$ es ta tipificación con el promedio y la desviación estándar ($x_i^*=\cfrac{x_i-\bar{X}}{s}$)

## Volvamos al problema

Como se comentó, tenemos que calcular una matriz $U$ para hacer la transformación de $X$ a $Y$. Sin embargo, hay que cumplir algunas condiciones:

1. Primero deben aparecer variables que representen la mayor varianza de las variables $X$.

2. No puede haber correlación entre las variables en $Y$.

Con esta información iniciemos el calculo de cada variable en $Y$, las llamaremos componentes:
<!DOCTYPE html>
$$\begin{bmatrix}Y_{11}\\Y_{12}\\\vdots \\Y_{1n}\end{bmatrix}=\begin{bmatrix}X_{11}& X_{21}& \cdots&X_{q1}\\X_{12}& X_{22}& \cdots&X_{q2}\\\vdots& \vdots &\ddots&\vdots\\ X_{1n}& X_{2n}& \cdots&X_{qn}\end{bmatrix}\begin{bmatrix}u_{11}\\u_{12}\\\vdots \\u_{1q}\end{bmatrix}$$




**Cálculo de la primera componente y Varianza Maxima:**

Calcular la varianza de la primera componente equivale a:
$$V(Y_1)=\frac{1}{n}Y^T\cdot Y=\frac{1}{n}u_1^TX^TXu_1=u_1^TVu_1$$

Luego, debemos maximizar:

$$V(Y_1)=u_1^TVu_1 \text{ \ \ sujeto a \ \ } u^Tu=1.$$

Es un problema de Lagrange, tenemos:

<!DOCTYPE html>
$$L=u_1^TVu_1-\lambda(u_1^Tu_1-1)$$

Derivando llegamos a:

$$(V-\lambda I)u_1=0$$

Que es un problema de valores y vectores propios.

Vamos a considerar el dataframe `Delitos_Colombia.csv` que contiene el número de delitos por departamento, en distintas categorías. Toda la información es numérica.

```python
plt.rcParams['figure.figsize'] = (15, 6)
plt.style.use('ggplot')
url_base = "https://raw.githubusercontent.com/lacamposm/Metodos_Estadisticos/main/data/"
df_name2 = "Delitos_Colombia.csv"
df2 = pd.read_csv(url_base + df_name2, sep=";", decimal=",", index_col=0)
df2
```

```python
# Seleccionamos las cuantitativas (ACP solo trabaja con cuantitativas). Cuando sea el caso seleccionamos
# así: -----> cuanti = df2.select_dtypes(np.number)
```

```python
sns.heatmap(df2.corr(), annot = True, cmap = "PuBu")
plt.show()
```

Para realizar un mejor análisis vamos considerar la siguiente tasa (para todas las variables):

$$Tasa Delitos Sexuales=100.000\times \frac{Delitos Sexuales}{Poblacion}$$

la cual representa la incidencia por cada $100.000$ habitantes.

```python
df2.div(df2['Poblacion'],axis=0).drop('Poblacion',axis=1)*100000
```

```python
df2["DSex_P"] = 100000*df2["Delitos_Sexuales"]/df2["Poblacion"]
df2["DHom_p"] = 100000*df2["Homicidios"]/df2["Poblacion"]
df2["DIntra_p"] = 100000*df2["Intrafamiliar"]/df2["Poblacion"]
df2["DTransi_p"] = 100000*df2["Transito"]/df2["Poblacion"]
df2["DAsal_p"] = 100000*df2["Asalto"]/df2["Poblacion"]
df2 = df2[["DSex_P", "DHom_p", "DIntra_p", "DTransi_p", "DAsal_p"]]
df2
```

```python
sns.heatmap(df2.corr(), annot = True, cmap = "PuBu")
plt.show()
```

```python
scaler2 = StandardScaler()        # Cargando el escalador estandar
scaler2.fit(df2)                  # Calcula las medias y las desviaciones
X_scaled = scaler2.transform(df2) # Estandariza la data.
pd.DataFrame(X_scaled, index = df2.index, columns = df2.columns)
```

```python
## Realizamos la rutina habitual.
pca2 = PCA(n_components = 5)
pca2.fit(X_scaled)
df2_pca = pca2.transform(X_scaled)
pca2_trans =  pd.DataFrame(df2_pca, index = df2.index, columns = ["PC1","PC2","PC3","PC4","PC5"])
pca2_trans
```

```python
expl = pca2.explained_variance_ratio_
expl
```

```python
print("Se ajustó un PCA a una data de tamaño:",df2.shape)
print("Se obtuvo un ajuste por PCA de tamaño:",pca2_trans.shape)
expl = pca2.explained_variance_ratio_
print("Cantidad de varianza explicada: {:.3f}%".format(sum(expl[0:2])*100))
```
