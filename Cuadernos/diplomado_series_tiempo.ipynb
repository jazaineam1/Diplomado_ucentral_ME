## ¿Qué se puede pronosticar?

Hay muchos fenómenos o variables que se pueden pronosticar; p. ej. los insumos que va a requerir una planta el siguiente mes o semana, las ventas de cierta sucursal en el mediano o largo plazo, la demanda de energía para una región del país en las siguientes 24 horas, etc.

Asi, podemos necesitar pronósticos para muy corto plazo (minutos, horas,...), diarios, semanales, mensuales, hasta de largo plazo (para los siguientes 5, 10 años).

Es importante saber que existen variables que se pueden pronosticar con gran exactitud (p. ej., la puesta del sol el día de mañana, o el siguiente eclipse lunar), y otras que, por mas que intentemos, nuestro pronostico sera un mero experimento aleatorio (pronosticar los números ganadores de la rifa del avión presidencial).

Entonces, que tan predecible es una variable o fenómeno depende de varias cosas:

1. Que tan bien conocemos los factores o elementos que influyen en el.
2. Que tantos datos tenemos disponibles.
3. Si el pronostico que vamos a realizar puede influir en el resultado de la variable a predecir.

Un caso de un tipo de series que son difíciles de pronosticar, son los tipos de cambio. Estos solo cumplen una de las tres características: tenemos muchos datos disponibles de ellos. Pero, desafortunadamente, hasta el momento no conocemos bien todo lo que afecta al tipo de cambio. Aparte, si se realiza algún pronostico de que, por ejemplo, el tipo de cambio va a aumentar, este puede influir en el valor futuro del mismo, logrando una *"profecia autocumplida"*. Esto es:

* Debido a las leyes de la oferta y la demanda, y a las nuevas expectativas de la gente a que va a subir el precio del dolar, la gente precavida comenzara a comprar dolares mientras estan "baratos".
* Esta nueva demanda ejercera presion en el tipo de cambio y comenzara a aumentar.
* Al ver esto las demas personas, veran que se esta cumpliendo el pronostico y querran adquirir mas dolares, antes de que suba tanto el precio.
* Esto podria provocar un efecto de bola de nieve y presionar mas el tipo de cambio.

Un buen pronostico captura los patrones genuinos de la variable y las relaciones existentes en los datos historicos, pero **no replica** eventos pasados que no volveran a suceder. Otra cualidad de un buen modelo de pronostico es aquel que captura la forma en que las cosas estan cambiando.

## Pronosticos, planeacion y metas

En los negocios, realizar pronosticos es una tarea estadistica muy comun; p. ej.

* Las plantas ensambladoras requieren realizar pronosticos de la demanda, para agendar la cantidad de obreros a requerir por turno.

* Una sucursal de carniceria debe prever cuantos kilogramos de carne se van a demandar al dia siguiente, para no quedarse sin existencias, ni con demasiados sobrantes.

* Una empresa de outsourcing de call center debe pronosticar la cantidad de llamadas que recibira, para habilitar ejecutivos de atencion a clientes suficientes.

* etc.

Sin embargo, los pronosticos de negocios a menudo se realizan muy deficientemente y son confundidos con la planeacion y las metas de la empresa.

* Los **pronosticos** se tratan de predecir el futuro de la manera mas certera posible, dada la informacion disponible, incluyendo datos historicos y conocimiento de eventos futuros que puedan impactar el pronostico.

* Los **objetivos** es lo que la empresa quiere que suceda (p. ej. crecer las ventas el 80%, *YOY*). Los objetivos o metas deberian estar ligados al pronostico, pero no siempre suceden. Seguido las metas se establecen sin ningun plan para como lograrlo o sin un pronostico de si son asequibles o no.

* La **planeacion** es la respuesta a los pronosticos y metas. Involucra llevar a cabo las acciones necesarias para que los pronosticos igualen a las metas.

Depende de la aplicacion especifica, pero las empresas necesitan pronosticos de corto, mediano y largo plazo.

* Los pronosticos de **corto plazo** se requieren para la programacion de personal, produccion y transporte, principalmente. Para ello, pronosticos de la demanda son requeridos de igual manera.

* Los pronosticos de **mediano plazo** se utilizan para determinar los requerimientos futuros de recursos, para poder adquirir materia prima, contratacion de personal, compra de maquinaria y equipo, etc.

* Los pronosticos de **largo plazo** son utilizados en la *planeacion estrategica*. Tales decisiones deben tomar en cuenta oportunidades de mercado y factores internos y externos de la empresa.

## Los pasos para llevar a cabo un pronostico

1. **La definicion del problema**. Se necesita tener un buen entendimiento del modo en que se usara el pronostico, quien lo va a consumir y como encaja dentro de la organizacion.

2. **Recopilacion de informacion**. Por lo general se requieren dos tipo de informacion: datos estadisticos y el *expertise* acumulado de la gente que recopila la informacion y usa los pronosticos.

3. **Analisis exploratorio de los datos**. Siempre se debe comenzar graficando los datos. ¿Se muestran patrones relevantes?, ¿existe una tendencia significativa?, ¿la estacionalidad es importante?, ¿existe evidencia de ciclos de negocio?, ¿hay datos *outliers* que se requieran explicar?

4. **Seleccion y estimacion de modelos**. La seleccion del modelo dependera de la disponibilidad de la informacion y del uso que se le dara al pronostico. Es comun comparar entre dos o tres modelos potenciales.

5. **Uso y evaluacion del modelo de pronostico**. Una vez seleccionado el modelo y estimado sus parametros, se realiza el pronostico. La evaluacion del desempenio del modelo solo se puede evaluar hasta que se obtiene la informacion real del periodo pronosticado.

### Re-muestreo
El re-muestreo implica cambiar la frecuencia de los datos:

**Upsampling:** de una frecuencia más baja a una más alta (diaria → por hora).

**Downsampling:** de una frecuencia más alta a una más baja (diaria → semanal).

Los datos de frecuencia más baja suelen agregarse: el valor para la semana X es la suma/promedio/máximo/mínimo/... de todos los valores diarios en la semana X.

Pasos del re-muestreo descendente:
`.resample(rule='M'|'W'|'D'|..)`

`aggregate` `.mean()`, `.sum()`

````python
# Resample the data to the average monthly values
data_monthly = data.resample("M").mean()

# Print the head of data
display(data.head())

# Print the head of resampled data
data_monthly.head()
````

# **Autocorrelación**

Para hacer predicciones, necesitamos saber cómo los valores futuros de una serie temporal dependen de los valores pasados.

La correlación de dos _variables aleatorias_ es una medida de la dependencia de una variable sobre la otra. Para las series temporales, la *auto*correlación, como el nombre podría sugerir, medirá la autocorrelación, es decir, la dependencia de una serie temporal sobre sí misma.

## **Retardos (Lags)**

## **Función de autocorrelación**
Función del retardo $\mathbf{n}$.

$$\large{Autocorr(\mathbf{n}) = corr(series, \,serie\,con\,retardo\,\mathbf{n})}$$

Mide cuánto depende el valor de la serie en el presente del valor hace $\mathbf{n}$ días.

`statsmodels` ofrece dos funciones relacionadas con la autocorrelación:
- `statsmodels.tsa.stattools.acf()` calculará la ACF para una serie temporal que se le pase, con un rango de 40 retardos de forma predeterminada.
- `statsmodels.graphics.tsaplots.plot_acf()` simplemente trazará la ACF, que es generalmente lo que deseamos.

# **Estacionariedad**

La mayoría de los modelos de series temporales asumen que la serie temporal es _estacionaria_. En lenguaje sencillo, esto significa que las características de la serie no dependen del _tiempo_ en que ocurren (pero podrían depender de los _valores_ pasados). Esto significa, entre otras cosas:

- Media constante
- Varianza constante

**Las series temporales que presentan tendencia y estacionalidad no son estacionarias.**

La mayoría de los modelos de series temporales asumen la estacionariedad de la serie que están modelando, por lo que a menudo debes eliminar la tendencia y la estacionalidad de la serie, construir el modelo y luego reintegrarlas.
## **Eliminación de Tendencia**

Existen algunas formas de eliminar la tendencia:

- _Ajustar primero la curva de tendencia_

- _Diferenciación_

En lugar de observar los valores reales de la serie, observa las _diferencias_ en los valores:

$U_t = Y_{t} - Y_{t-1}$

Aquí tienes un ejemplo:

<br>

| fecha       | serie original | serie diferenciada |
|-------------|----------------|--------------------|
| 2020-05-01  | 5              | NaN                |
| 2020-05-02  | 7              | 7 - 5 = **2**      |
| 2020-05-03  | 8              | 8 - 7 = **1**      |
| 2020-05-04  | 8              | 8 - 8 = **0**      |
| 2020-05-06  | 9              | 9 - 8 = **1**      |

<br>

La serie original está aumentando, pero las diferencias no lo están.

El método `.diff()` de `pandas` se puede utilizar para diferenciar la serie.

- _Otras transformaciones_

Los rendimientos y los logaritmos de los rendimientos, $\log{\frac{Y_t - Y_{t-1}}{Y_t}}$, también se utilizan comúnmente en la modelización de precios de acciones.

---
Utilicemos la diferenciación para eliminar la tendencia de la serie de producción de gasolina.

````python
# Difference data_monthly
data_monthly_diff=data_monthly.diff()
display(data_monthly.tail())
display(data_monthly_diff.head())
# Show differenced values
data_monthly_diff.plot()
````

## **Lidiando con la estacionalidad**

Lidiar con la estacionalidad es mucho más complicado. Muchas organizaciones (por ejemplo, la Oficina del Censo de EE. UU.) han desarrollado sus propios métodos para eliminar la estacionalidad de los datos.

Algunas cosas que puedes hacer por ti mismo:

- _Modelar dentro de un período_

  Si solo estás utilizando datos de 14 días en junio, ¿importa si hay un pico estacional anual alrededor de las vacaciones de invierno?..

- _Usar datos ajustados estacionalmente_

  Por ejemplo, la mayoría de las agencias gubernamentales de EE. UU. proporcionan datos ajustados estacionalmente de forma predeterminada.

- _Realizar el ajuste estacional tú mismo_

  - Usar diferenciación de orden superior
  - Usar los residuos de `seasonal_decompose()`, que están contenidos en el atributo `.resid`
  
- _Usar modelos especiales como el Modelo ARIMA Estacional (SARIMA)_

Aprende sobre la [prueba de Dickey-Fuller](https://en.wikipedia.org/wiki/Dickey%E2%80%93Fuller_test) para la estacionariedad y [`statsmodels.tsa.stattools.adfuller()`](https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html)</font>

```python
import statsmodels.api as sm

def prueba_dickey_fuller(x, maxlag=None, regression='c', autolag='AIC', store=False, regresults=False):
    """
    Prueba de raíz unitaria Dickey-Fuller aumentada.

    La prueba de Dickey-Fuller aumentada se puede utilizar para probar una raíz unitaria en un proceso univariado en presencia de correlación serial.

    Parámetros
    ----------
    x : array_like, 1d
        La serie de datos a probar.

    maxlag : {None, int}, opcional
        Lags máximos que se incluyen en la prueba. Se utiliza el valor predeterminado de 12*(nobs/100)^{1/4} cuando es None.

    regression : {"c","ct","ctt","n"}, opcional
        Orden constante y de tendencia para incluir en la regresión.
        
        "c" : solo constante (predeterminado).
        
        "ct" : constante y tendencia.
        
        "ctt" : constante, y tendencia lineal y cuadrática.
        
        "n" : sin constante, sin tendencia.

    autolag : {"AIC", "BIC", "t-stat", None}, opcional
        Método a utilizar al determinar automáticamente la longitud del rezago entre los valores 0, 1, ..., maxlag.
        
        Si es "AIC" (predeterminado) o "BIC", entonces se elige el número de rezagos para minimizar el criterio de información correspondiente.
        
        Elección basada en "t-stat" del maxlag. Comienza con maxlag y descarta un rezago hasta que la estadística t en el último valor de rezago sea significativa utilizando una prueba al 5%.
        
        Si es None, entonces el número de rezagos incluidos se establece en maxlag.

    store : bool, opcional
        Si es True, se devuelve una instancia de resultados además de la estadística adf. El valor predeterminado es False.

    regresults : bool, opcional
        Si es True, se devuelven los resultados completos de la regresión. El valor predeterminado es False.
    """
    return sm.tsa.adfuller(x, maxlag=maxlag, regression=regression, autolag=autolag, store=store, regresults=regresults)
```
**Notas:**

La hipótesis nula de la prueba de Dickey-Fuller aumentada es que hay una raíz unitaria, con la alternativa de que no hay raíz unitaria. Si el valor p está por encima de un tamaño crítico, entonces no podemos rechazar que hay una raíz unitaria.

Los valores p se obtienen a través de una aproximación de superficie de regresión de MacKinnon 1994, pero utilizando las tablas actualizadas de 2010. Si el valor p está cerca de ser significativo, entonces se deben utilizar los valores críticos para juzgar si rechazar la hipótesis nula.

La opción autolag y maxlag para ella se describen en Greene.

Consulta el cuaderno "Estacionariedad y detrending (ADF/KPSS)" para obtener una descripción general.

**Returns:**

- `adf` : float
  La estadística de prueba.

- `pvalue` : float
  Valor p aproximado de MacKinnon basado en MacKinnon (1994, 2010).

- `usedlag` : int
  Número de rezagos utilizados.

- `nobs` : int
  Número de observaciones utilizadas para la regresión ADF y el cálculo de los valores críticos.

- `critical_values` : dict
  Valores críticos para la estadística de prueba en los niveles 1 %, 5 % y 10 %. Basado en MacKinnon (2010).

- `icbest` : float
  Criterio de información maximizado si autolag no es None.

- `resstore` : ResultStore, opcional
  Una clase ficticia con resultados adjuntos como atributos.

````python
#realice la prueba de Dickey - Fuller
from statsmodels.tsa.stattools import adfuller
result=adfuller(data_monthly)

# Imprimir los resultados de la prueba
print('Resultado de la prueba Dickey-Fuller:')
print(f'Estadística de prueba: {result[0]}')
print(f'Valor p: {result[1]}')
print('Valores críticos:')
for key, value in result[4].items():
    print(f'   {key}: {value}')
print(f'Estimación del número de rezagos utilizados: {result[2]}')
````

# **Modelos de series temporales**

Esta sección contendrá algunas fórmulas matemáticas, pero no te preocupes: ¡las intuiciones son bastante simples! Si no comprendes completamente las fórmulas, ¡eso no debería detenerte para modelar!

## **Dos modelos básicos**

#### **1. Ruido blanco**

> El mejor modelo para un valor futuro es simplemente una media constante.

$$\large{valor\_hoy = media + ruido\_hoy}$$

$$Y_t = \mu + \varepsilon_t$$

Esta serie temporal simplemente oscila alrededor de la media:

![Ruido blanco](https://github.com/datacamp/time-series-analysis-in-python-live-training/raw/master/assets/white_noise.png)

#### **2. Camino aleatorio**

> El mejor modelo para un valor futuro es el valor anterior.

$$\large{valor\_hoy = valor\_ayer + ruido\_hoy}$$
<br>

$$Y_t = Y_{t-1} + \varepsilon_t$$

Esta serie temporal deriva aleatoriamente:

![Camino aleatorio](https://github.com/datacamp/time-series-analysis-in-python-live-training/raw/master/assets/random_walk.png)

## **Modelo de promedio móvil (MA)**

Los choques/ruidos/errores de hace unos días todavía pueden afectar la serie hoy: por ejemplo, un gran shock en el mercado puede persistir durante días. Podemos mejorar ligeramente el modelo de ruido blanco teniendo en cuenta los errores previos.

$$MA(q): \large{valor\_hoy = media + promedio(pasado \, \mathrm{q} \,ruidos)}$$

$$Y_t = \mu + \varepsilon_t + \phi_1 \varepsilon_{t-1} + \phi_2 \varepsilon_{t-2} +...+ \phi_q \varepsilon_{t-q} $$

"Promedio" significa "promedio ponderado", con pesos $\phi_i$ posiblemente disminuyendo a medida que avanzamos en el futuro.

### **Detección del orden del modelo MA**

El modelo $MA(q)$ solo contiene información sobre los últimos $q$ períodos de tiempo, por lo que no debería haber una autocorrelación significativa después de $q$ rezagos.

En consecuencia, los modelos $MA(q)$ son adecuados para modelar series temporales con una función de autocorrelación (ACF) que "se detiene" en un cierto punto.

Por ejemplo, esto es generado a partir de un proceso $MA(2)$ real:

![ACF de residuos](https://github.com/datacamp/time-series-analysis-in-python-live-training/raw/master/assets/ma.png)

## **Modelo autoregresivo (AR)**

En el modelo autoregresivo (AR), el valor futuro no se basa solo en un único valor pasado, sino en varios valores pasados (en lugar de los _residuos_ pasados, como en el modelo $MA(q)$):

<br>
$$AR(p): \large{valor\_hoy = media + promedio(valores \, pasados \, \mathrm{p}) + ruido\_hoy}$$

$$Y_t = \mu + \theta_1 Y_{t-1} + \theta_2 Y_{t-2} + \dots + \theta_p Y_{t-p} + \varepsilon_t$$


donde el promedio es nuevamente un promedio ponderado. Hasta qué punto en el pasado se extiende el modelo, es decir, el valor de $p$, es el _orden_ del modelo. El modelo generalmente se denota como AR(p).

Los órdenes bajos son los más comunes: $AR(1)$, $AR(2)$,...

$$\text{AR(1): }Y_t = \mu + \theta Y_{t-1} + \varepsilon_t$$


Cuando $\theta = 1$, esto es un modelo de _camino aleatorio_. Cuando $\theta \geq 1$, el proceso no es estacionario: la varianza aumenta.

### **Detección del orden del modelo**

A diferencia del modelo $MA(q)$, en los modelos $AR(p)$ de cualquier orden, el valor actual depende de _todos_ los valores pasados:


$$
\begin{align}
Y_t = \mu + \theta \color{red}{Y_{t-1}} + \varepsilon_t &= \mu + \theta \color{red}{(\mu + \theta {Y_{t-2}} + \varepsilon_{t-1})} + \varepsilon_t\\
\\
\mu + \theta \color{red}{(\mu + \theta \color{blue}{Y_{t-2}} + \varepsilon_{t-1})} + \varepsilon_t &=\mu + \theta \color{red}{(\mu + \theta \color{blue}{(\mu + \theta {Y_{t-3}} + \varepsilon_{t-2})} + \varepsilon_{t-1})} + \varepsilon_t = \\
\\
\mu + \theta \color{red}{(\mu + \theta \color{blue}{(\mu + \theta \color{magenta}{{Y_{t-3}}} + \varepsilon_{t-2})} + \varepsilon_{t-1})} + \varepsilon_t &= \dots
\end{align}
$$


Cuando $|\theta_i| < 1$, la dependencia disminuirá a medida que avanzamos cada vez más en el pasado:


$$AR(1) : Y_{t} = constante + \theta Y_{t-1} + \theta^2 Y_{t-2} + \theta^3 Y_{t-3}  + \dots $$


Entonces, la función de autocorrelación no "se detiene" &mdash; simplemente disminuye a 0.

Aquí tienes un ejemplo de un proceso $AR(3)$ real:

![Proceso AR(3)](https://github.com/datacamp/time-series-analysis-in-python-live-training/raw/master/assets/ar.png)

````python
# Plot ACF for unemployment_diff
unemployment_diff=unemployment.diff().dropna()
plot_acf(unemployment_diff);
````

El valor de la autocorrelación oscila entre negativo y positivo, pero su magnitud disminuye.

#### **Función de autocorrelación parcial**

La función de autocorrelación parcial es una versión modificada de la función de autocorrelación que tiene en cuenta las dependencias entre los rezagos anteriores y elimina las dependencias superfluas que se muestran en la función de correlación.

- `statsmodels.tsa.stattools.pacf()` devolverá los valores de la PACF.
- `statsmodels.graphics.tsaplots.plot_pacf()` mostrará la gráfica de la PACF.

Para las series temporales que se comportan como un proceso $AR(p)$, la PACF "se detiene" en cierto rezago. Ese rezago puede servir como el orden del modelo.

Aquí tienes la PACF de un verdadero modelo $AR(2)$:

![PACF de modelo AR(2)](https://github.com/datacamp/time-series-analysis-in-python-live-training/raw/master/assets/pacf.png)

````python
# Import plot_pacf
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(unemployment_diff);
````

A diferencia de la gráfica de la ACF, no hay una disminución gradual después del rezago 1.

````python

# Realizar predicciones trimestrales a fin de mes
forecast_steps = 27  # Número de pasos de predicción hacia el futuro
last_date = time_series.index[-1]
last_month = last_date.to_period("M")  # Convertir la última fecha a periodo mensual
forecast_dates = pd.period_range(start=last_month + 1, periods=forecast_steps, freq='Q').to_timestamp(how="E")  # Fechas trimestrales a fin de mes

# Realizar predicciones
forecast, conf_int = model.predict(n_periods=forecast_steps, return_conf_int=True)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Datos reales')
plt.plot(forecast_dates, forecast, color='red', label='Predicciones')
plt.fill_between(forecast_dates, conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.3)
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Predicciones con modelo ARIMA')
plt.legend()
plt.show()
````

|Proceso| Función de autocorrelación (ACF) |Función de autocorrelación parcial  (ACFP)|
|---|---|---|
|MA(q)|Solo los q primeros coeficientes son significativos. El resto se anulan bruscamente (coef. 0 para retardo >q)| Decrecimiento rápido exponencial atenuado u ondas sinusoidales.|
|AR(p)| Decrecimiento rápido exponencial atenuado u ondas sinusoidales. |Solo los p primeros coeficientes son significativos. El resto se anulan bruscamente (coef. 0 para retardo >q)|
|ARIMA(p, d, q)|Comportamiento irregular en los retardos(1, ... , q) con q picos. Decrecimiento para retardos posteriores a q.|Decrece (aproximadamente con exponenciales atenuados y ondas sinusoidales). No cero pronto.|
